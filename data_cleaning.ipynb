```python
import pandas as pd

df=pd.read_csv("indeed_jobs.csv")

#salary parsing
df["monthly"]=df["Salary Estimate"].apply(lambda x: 1 if "a month" in x.lower() else 0)
df["hourly"]=df["Salary Estimate"].apply(lambda x: 1 if "an hour" in x.lower() else 0)
df["daily"]=df["Salary Estimate"].apply(lambda x: 1 if "a day" in x.lower() else 0)


df=df[df["Salary Estimate"]!="-1"]
salary=df["Salary Estimate"]
minus_inr=salary.apply(lambda x:x.replace("₹",""))
minus_inr

min_mth=minus_inr.apply(lambda x: x.lower().replace("a month","").replace("an hour","").replace("a day","").replace("a year",""))
# min_mth
min_mth=min_mth.apply(lambda x: x.replace(",",""))

df["min_salary"]=min_mth.apply(lambda x:int(x.split(" ")[1] if "from" in x else(x.split("-")[0] if "-" in x else 0)))
df["max_salary"]=min_mth.apply(lambda x:int( x.split("up to")[1] if "up to" in x else(x.split("-")[1] if "-" in x else 0)))
df["avg_salary"]=(df.min_salary+df.max_salary)/2

#state field
df["Location"]=df["Location"].apply(lambda x:x.split("\r")[0])
# df["Location"]=df["Location"].apply(lambda x:x.replace(" ",""))
df["job_state"]=df["Location"].apply(lambda x:x.split(",")[1] if "," in x else x)
df["job_state"]=df["job_state"].apply(lambda x:x.replace(" ",""))



df["Benefits and perks"]=df["Benefits and perks"].apply(lambda x: x.replace("'","").replace("[","").replace("]",""))
df["Job Type"]=df["Job Type"].apply(lambda x: str(x).replace("'","").replace("[","").replace("]",""))


df["Job Description"]=df["Job Description"].apply(lambda x: x.split("Full Job Description")[1] if "Full Job Description" in x else x)




#python
df["python_yn"]=df["Job Description"].apply(lambda x:1 if "python" in x.lower() else 0)
# df["python_yn"].value_counts()

#r studio
df["rstudio_yn"]=df["Job Description"].apply(lambda x:1 if "r studio" in x.lower() or "r-studio" in x.lower() else 0)
# df["rstudio_yn"].value_counts()

#spark
df["spark_yn"]=df["Job Description"].apply(lambda x:1 if "spark" in x.lower() in x.lower() else 0)
# df["spark_yn"].value_counts()

#aws
df["aws_yn"]=df["Job Description"].apply(lambda x:1 if "aws" in x.lower() else 0)
# df["aws_yn"].value_counts()

#excel
df["excel_yn"]=df["Job Description"].apply(lambda x:1 if "excel" in x.lower() else 0)


# df["excel_yn"].value_counts()
# df.job_state.value_counts()
# min_mth
df.to_csv("Salary data cleaned.csv",index=False)


pd.read_csv("Salary data cleaned.csv")
```
